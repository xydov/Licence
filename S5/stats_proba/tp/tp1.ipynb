```python
def loi_normale_0(mu, sigma, pas):

    # Normalisation
    mini = -1
    maxi = 6
    
    # mu = location | sigma = scale
    # Loi theorique
    vec = np.arange(mini, maxi, pas)
    xt = loiT.norm.pdf(vec, mu, sigma)

    # Affichage
    plt.figure()
    plt.plot(vec, xt, 'r', label='resultat theorique')
    plt.title('Loi normale')
    plt.xlabel('Intervalles')
    plt.ylabel('Probabilites')
    plt.legend()
    plt.show()
```

```python
loi_normale_0(2, 1, 0.001)
```

```python
# Question 2
def loi_normale_expo_0(mu, sigma, lambd, pas):

    # Normalisation
    mini = 0
    maxi = 10
    
    # mu = location | sigma = scale
    # Loi theorique
    vec = np.arange(mini, maxi + 1, pas)
    xnorm = loiT.norm.pdf(vec, mu, sigma)
    xexp = loiT.expon.pdf(vec)
    # , 1/lambd, 1/pow(lambd, 2)

    # Affichage
    plt.figure()
    plt.plot(vec, xnorm, 'r', label='resultat theorique normale')
    plt.plot(vec, xexp, 'g', label='resultat theorique exponentielle')
    plt.title('Loi normale et exponentielle')
    plt.xlabel('Intervalles')
    plt.ylabel('Probabilites')
    plt.legend()
    plt.show()
```

# EXERCICE 1 : comparaison de lois uniformes : theorique & pratique

```python
def loi_uniforme(nb_val, p1, p2, barres, pas): # A COMPLETER...

    # Loi pratique (valeurs aleatoires)
    xp = loi.uniform(p1, p2, size=nb_val)

    # Normalisation
    mini = p1
    maxi = p2
    # Loi theorique
    vec = np.arange(mini, maxi, pas)
    xt = loiT.uniform.pdf(vec, loc=p1, scale=p2 - p1)

    # Affichage
    plt.figure()
    plt.hist(xp, barres, density=True, label='resultat pratique')
    plt.plot(vec, xt, 'r', label='resultat theorique') # A MODIFIER dans le cas discret par 'or'
    plt.title('Loi uniforme')
    plt.xlabel('Intervalles')
    plt.ylabel('Probabilites')
    plt.legend() 
    plt.show()
```

```python
# DEBUT DU PROGRAMME PRINCIPAL

# Constante
nb_barres = 20
pas_reel = 0.02
pas_discret = 1
```

```python
# (a) Tests de la loi Uniforme : loi discrete ou reelle
# 50 valeurs
loi_uniforme(50, 0, 20, nb_barres, pas_reel)
# 10000 valeurs
loi_uniforme(10000, 0, 20, nb_barres, pas_reel)
# 10000 valeurs
loi_uniforme(10000, -5, 5, nb_barres, pas_reel)
```

```python
def loi_exponentielle(nb_val, lambd, barres, pas):

    # Loi pratique (valeurs aleatoires)
    xp = loi.exponential(1/lambd, size=nb_val)

    # Normalisation
    mini = min(xp)
    maxi = max(xp)
    
    # Loi theorique
    vec = np.arange(mini, maxi, pas)
    xt = loiT.expon.pdf(vec, loc=0, scale=1/lambd)
    

    # Affichage
    plt.figure()
    plt.hist(xp, barres, density=True, label='resultat pratique')
    plt.plot(vec, xt, 'r', label='resultat theorique')
    plt.title('Loi exponentielle')
    plt.xlabel('Intervalles')
    plt.ylabel('Probabilites')
    plt.legend() 
    plt.show()
```

```python
# (b) Tests de la loi Exponentielle
# 50 valeurs
loi_exponentielle(50, 0.02, nb_barres, pas_reel)
# 10000 valeurs
loi_exponentielle(10000, 0.02, nb_barres, pas_reel)
# 10000 valeurs
loi_exponentielle(10000, 0.8, nb_barres, pas_reel)
```

```python
def loi_geometrique(nb_val, p, barres, pas):

    # Loi pratique (valeurs aleatoires)
    xp = loi.geometric(p, size=nb_val)

    # Normalisation
    mini = min(xp)
    maxi = max(xp)
    
    # Loi theorique
    vec = np.arange(mini, maxi, pas)
    xt = loiT.geom.pmf(vec, p=p)
    

    # Affichage
    plt.figure()
    plt.hist(xp, barres, density=True, label='resultat pratique')
    plt.plot(vec, xt, 'or', label='resultat theorique')
    plt.title('Loi géométrique')
    plt.xlabel('Intervalles')
    plt.ylabel('Probabilites')
    plt.legend() 
    plt.show()
```

```python
# (c) Tests de la loi geometrique 
# 50 valeurs 
loi_geometrique(50, 0.07, nb_barres, pas_discret)
# 10000 valeurs
loi_geometrique(10000, 0.07, nb_barres, pas_discret)
# 10000 valeurs
loi_geometrique(10000, 0.2, nb_barres, pas_discret)
```

```python
def loi_poisson(nb_val, p, barres, pas):

    # Loi pratique (valeurs aleatoires)
    xp = loi.poisson(p, size=nb_val)

    # Normalisation
    mini = min(xp)
    maxi = max(xp)
    
    # Loi theorique
    vec = np.arange(mini, maxi, pas)
    xt = loiT.poisson.pmf(vec, p)
    

    # Affichage
    plt.figure()
    plt.hist(xp, barres, density=True, label='resultat pratique')
    plt.plot(vec, xt, 'or', label='resultat theorique')
    plt.title('Loi géométrique')
    plt.xlabel('Intervalles')
    plt.ylabel('Probabilites')
    plt.legend() 
    plt.show()
```

```python
# (d) Tests de la loi poisson
# 50 valeurs 
loi_poisson(50, 5, nb_barres, pas_discret)
# 10000 valeurs
loi_poisson(10000, 5, nb_barres, pas_discret)
# 10000 valeurs
loi_poisson(10000, 0.5, nb_barres, pas_discret)
# 10000 v
loi_poisson(10000, 50, nb_barres, pas_discret)
```

```python
def loi_normale(nb_val, mu, sigma, barres, pas):

    # Loi pratique (valeurs aleatoires)
    xp = loi.normal(mu, sigma, size=nb_val)

    # Normalisation
    mini = min(xp)
    maxi = max(xp)
    
    # Loi theorique
    vec = np.arange(mini, maxi, pas)    
    xt = loiT.norm.pdf(vec, mu, sigma)
    

    # Affichage
    plt.figure()
    plt.hist(xp, barres, density=True, label='resultat pratique')
    plt.plot(vec, xt, 'r', label='resultat theorique')
    plt.title('Loi géométrique')
    plt.xlabel('Intervalles')
    plt.ylabel('Probabilites')
    plt.legend() 
    plt.show()
```

```python
# (e) Tests de la loi normale
# 50 valeurs 
loi_normale(50, 0, 1, nb_barres, pas_reel)
# 10000 valeurs
loi_normale(10000, 0, 1, nb_barres, pas_reel)
# 10000 valeurs
loi_normale(10000, 5, 0.5, nb_barres, pas_reel)
# 10000 
loi_normale(10000, 50, 500, nb_barres, pas_reel)
```
